```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/state_wise_crop_production.csv')
print(df.head())
```

```python
print(df.info())
```

# New Section

```python
print(df['Crop'].unique())
```

```python
print(df['State'].unique())
```

```python
print(df.columns)
```

```python
df.groupby(['State', 'Crop'])['Cost of Cultivation (`/Hectare) A2+FL'].sum().sort_values(ascending = True).plot(kind = 'barh',figsize = (13,13))
plt.tight_layout()
plt.title('Crop by State Based on Total Cost of Cultivation (`/Hectare) A2+FL')
plt.legend()
plt.show()
```

```python
df.groupby(['State', 'Crop'])['Cost of Production (`/Quintal) C2'].sum().sort_values(ascending = True).plot(kind = 'barh',figsize = (13,13))
plt.tight_layout()
plt.title('Crop by State Based on Total Cost of Production (`/Quintal) C2')
plt.legend()
plt.show()
```

```python
df.groupby(['State', 'Crop'])['Yield (Quintal/ Hectare) '].sum().sort_values(ascending = True).plot(kind = 'barh',figsize = (13,13))
plt.tight_layout()
plt.title('Crop by State Based on Total Yield (Quintal/ Hectare)')
plt.legend()
plt.show()
```
